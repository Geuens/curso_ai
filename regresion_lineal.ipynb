```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Crear el gráfico de dispersión
plt.scatter(X*1000, y, color='blue', label='Datos')

# Configuración del gráfico
plt.title('Gráfico de dispersión y línea de regresión')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Crear un modelo de regresión lineal
modelo_regresion = LinearRegression()

# Ajustar el modelo a los datos
modelo_regresion.fit(X*1000, y)

# Predecir y para x=1850
x_nuevo = np.array([[1850]])
y_predicho = modelo_regresion.predict(x_nuevo)

print("Predicción para x=1850:", y_predicho)
```

```python
# Representar el punto predicho en rojo
plt.scatter(X*1000, y, color='blue', label='Datos')
plt.scatter(x_nuevo, y_predicho, color='red', marker='o', label='Predicción para x=1850')



# Configuración del gráfico
plt.title('Gráfico de dispersión y línea de regresión')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
#Prediccion incidencias
print(y_predicho)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = modelo_regresion.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```
