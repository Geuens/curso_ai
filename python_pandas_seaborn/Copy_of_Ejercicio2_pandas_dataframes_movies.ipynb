```python
import pandas as pd
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1033_pgXO86pl9hwlIvGZZtNgmbiYJ7ps')
print(df.to_string(index=False))
```

```python
df['Rentabilidad'] = round(df['Ganancias'] / df['Presupuesto'] * 100, 2)
```

```python
import matplotlib.pyplot as plt

# Agrupando y calculando la media
df_grouped_year = df.groupby('Año de Emisión')['Recaudación'].mean()

# Crear un gráfico de línea
plt.figure(figsize=(10, 6))
df_grouped_year.plot(kind='line')
plt.title('Recaudación Media por Año')
plt.xlabel('Año de Emisión')
plt.ylabel('Recaudación Media')
plt.grid(True)
plt.show()

# Agrupando y calculando la media de rentabilidad por género
df_genre_profitability = df.groupby('Género')['Rentabilidad'].mean()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
df_genre_profitability.plot(kind='bar')
plt.title('Rentabilidad Promedio por Género')
plt.xlabel('Género')
plt.ylabel('Rentabilidad Promedio')
plt.xticks(rotation=45)  # Rota las etiquetas del eje X para mejor lectura
plt.grid(axis='y')
plt.show()
```

```python
# Agrupando y calculando la media de rentabilidad por género
df_genre_presupuesto = df.groupby('Género')['Presupuesto'].mean()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
df_genre_presupuesto.plot(kind='bar')
plt.title('Presupuesto Promedio por Género')
plt.xlabel('Género')
plt.ylabel('Presupuesto Promedio')
plt.xticks(rotation=45)  # Rota las etiquetas del eje X para mejor lectura
plt.grid(axis='y')
plt.show()
```

```python
#tipo de presupuesto

bins = [0, 1.900000e+08, 2.285000e+08, 2.525000e+08]
labels = ['Presupuesto Bajo', 'Presupuesto Medio', 'Presupuesto Alto']

df['Tipo Presupuesto'] = pd.cut(df['Presupuesto'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

```python
# Agrupando y calculando la media de rentabilidad por género
df_rentabilidad_tipo_presupuesto = df.groupby('Tipo Presupuesto')['Rentabilidad'].mean()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
df_rentabilidad_tipo_presupuesto.plot(kind='bar')
plt.title('Rentabilidad Promedio por tipo presupuesto')
plt.xlabel('Tipo')
plt.ylabel('Rentabilidad Promedio')
plt.xticks(rotation=45)  # Rota las etiquetas del eje X para mejor lectura
plt.grid(axis='y')
plt.show()
```

```python
# Agrupando y calculando la media de rentabilidad por género
df_rentabilidad_imdb = df.groupby('Calificación IMDB')['Rentabilidad'].mean()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
df_rentabilidad_imdb.plot(kind='line')
plt.title('Rentabilidad Promedio por tipo presupuesto')
plt.xlabel('Tipo')
plt.ylabel('Rentabilidad Promedio')
plt.xticks(rotation=45)  # Rota las etiquetas del eje X para mejor lectura
plt.grid(axis='y')
plt.show()
```
